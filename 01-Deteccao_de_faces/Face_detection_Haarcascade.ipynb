```python
# Importando bibliotecas;
import cv2

from matplotlib import pyplot as plt
```

```python
# Lendo Imagem;
img = cv2.imread("imagem01.jpg")
# Convertendo Imagem de BGR para RGB;
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
# Configurando plotagem das figuras;
plt.figure(figsize=(15,15)) # tamanho do plot
# subplots para forma uma unica imagem;
plt.subplot(1,2,1), plt.imshow(img);
plt.title("Imagem Original")
plt.subplot(1,2,2), plt.imshow(img_gray,"gray");
plt.title("Imagem escala de cinza");
```

```python
# Carregando detector treinado do XML;
face_detector = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

detector = face_detector.detectMultiScale(img_gray, scaleFactor=1.3, minSize=(30,30))
```

```python
# Desenhando box
for (x, y, l, a) in detector:
  #print(x, y, l, a)
  cv2.rectangle(img, (x, y), (x + l, y + a), (0,0,255), 2)
plt.figure(figsize=(15,15)) # tamanho do plot
plt.imshow(img);
```
