```python
# Bibliotecas
import cv2

from matplotlib import pyplot as plt
```

```python
# Lendo imagem
image = cv2.imread("bodys.png")
```

```python
# Carregando detector Haarcascade j treinado;
cascade_body = cv2.CascadeClassifier("haarcascade_fullbody.xml")

# Transformando imagem de BGR para Escala de Cinza;
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Realizando deteco de corpo;
detections = cascade_body.detectMultiScale(image_gray, scaleFactor=1.1, minSize=(50,50))

# Desenhando Bounding Box nos corpos detectados;
for (x, y, l, a) in detections:
  cv2.rectangle(image, (x, y), (x + l, y + a), (0,255,0), 2)

# transforma a imagem de BGR para RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plot da imagem;
plt.figure(figsize=(10,10))
plt.imshow(image);
plt.title("Corpos detectados: " + str(len(detections)));
```
